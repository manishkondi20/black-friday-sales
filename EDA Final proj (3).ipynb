#### Hypothesis Testing

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
male=[1180, 1200, 1150, 1210, 1220]
female=[1300, 1280, 1270, 1290, 1250]
```

```python
ttest,pvalue=stats.ttest_ind(male,female)
if pvalue<0.05:
    print('reject')
else:
    print('accept')
    
```

```python
sns.kdeplot(male,color='red')
sns.kdeplot(female,color='green')
plt.grid()
plt.title('purchase hypo')
plt.legend(['male','female'])
```
